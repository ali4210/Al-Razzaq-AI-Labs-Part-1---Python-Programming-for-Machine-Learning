```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(0, 1, 1000)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Generate random data
x = np.random.rand(100)
y = x + np.random.normal(0, 0.1, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', marker='o', alpha=0.5)
plt.title('Scatter Plot of Random Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='green', marker='x')

plt.title('Annotated Scatter Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Arbitrarily choosing an index for annotation
index = 10
plt.annotate('Special Point', xy=(x[index], y[index]),
             xytext=(x[index]+0.1, y[index]-0.1),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.grid(True)
plt.show()
```
