```python
from bs4 import BeautifulSoup
import requests

print("Libraries imported successfully!")
```

```python
url = "https://example.com"
response = requests.get(url)

# Verify the response status
if response.status_code == 200:
    print("Successfully fetched the webpage content!")
else:
    print("Failed to fetch the webpage content.")
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print(soup.prettify()[:500])  # Display the first 500 characters of the prettified HTML
```

```python
# Example: Extract and print all headers (h1) from the webpage
headers = soup.find_all('h1')
for header in headers:
    print(header.get_text())
```

```python
cleaned_data = [header.get_text().strip() for header in headers]
print(cleaned_data)
```
