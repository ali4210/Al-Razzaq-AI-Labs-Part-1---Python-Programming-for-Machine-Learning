```python
import pandas as pd
```

```python
# Creating a Series from a list
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
```

```python
# Creating a Series with custom indices
series_with_index = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])
print(series_with_index)
```

```python
# Creating a dictionary for DataFrame
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
```

```python
# Converting dictionary to DataFrame
df = pd.DataFrame(data_dict)
print(df)
```

```python
# Accessing 'Name' column
name_column = df['Name']
print(name_column)
```

```python
# Accessing the first row using loc
first_row = df.loc[0]
print(first_row)
```

```python
# Modifying age of the first record
df.at[0, 'Age'] = 26
print(df)
```

```python
# Adding a new column 'Country'
df['Country'] = ['USA', 'USA', 'USA']
print(df)
```
